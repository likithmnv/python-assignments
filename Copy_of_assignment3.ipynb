Write a Python program to check whether a list contains any positive values using any().

```python
numbers = []
n = int(input("Enter the number of elements:"))
for i in range(n):
  num = int(input("Enter element:"))
  numbers.append(num)
  result = any(num > 0 for num in numbers)
  print("List:", numbers)
  print("Contains any positive value?", result)
```

Write a Python program to check whether all numbers in a list are positive using all().

```python
numbers = []
n = int(input("Enter the number of elements:"))
for i in range(n):
  num = int(input("Enter element:"))
  numbers.append(num)
  result = all(num > 0 for num in numbers)
  print("List:", numbers)
  print("Are all numbers positive?", result)
```

Write a Python program that uses multiple built-in functions together to process a list of numbers.

```python
numbers = []
n = int(input("Enter the number of elements: "))
for i in range(n):
    num = int(input("Enter element: "))
    numbers.append(num)

print("List:", numbers)
print("Length of the list:", len(numbers))
print("Sum of elements:", sum(numbers))
print("Largest number:", max(numbers))
print("Smallest number:", min(numbers))
print("Are all numbers positive?", all(num > 0 for num in numbers))
print("Is there any even number?", any(num % 2 == 0 for num in numbers))
```
