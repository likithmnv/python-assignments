Calculate compound interest using math.pow().

```python
import math

principal = float(input("Enter the principal amount: "))
rate = float(input("Enter the annual interest rate (%): "))
time = float(input("Enter the time (years): "))

amount = principal * math.pow((1 + rate / 100), time)

compound_interest = amount - principal

print("Amount:", amount)
print("Compound interest:", compound_interest)
```

Find the area and circumference of a circle using math.pi.

```python
import math

radius = float(input("Enter the radius of the circle: "))

area = math.pi * radius * radius
circumference = 2 * math.pi* radius

print("Area of the circle: ", area)
print("Circumference of the cirlce: ", circumference)
```

Create a scientific calculator that supports square root, factorial, logarithm, power, trigonometric functions, GCD, and LCM.

```python
import math

print("===== Scientific Calculator =====")
print("1. Square Root")
print("2. Factorial")
print("3. Logarithm (Base 10)")
print("4. Power")
print("5. Trigonometric Functions")
print("6. GCD")
print("7. LCM")

while True:
  try:
    choice_input = input("Enter your choice (1-7): ")
    choice = int(choice_input)
    if 1 <= choice <= 7:
      break
    else:
      print("Invalid choice. Please enter a number between 1 and 7.")
  except ValueError:
    print("Invalid input. Please enter a number.")

if choice == 1:
  num = float(input("Enter a number:"))
  print("Square root =", math.sqrt(num))

elif choice == 2:
  num = int(input("Enter a number: "))
  print("Factorial =", math.factorial(num))

elif choice == 3:
  num = float(input("Enter a positive number: "))
  print("Logarithm =", math.log10(num))

elif choice == 4:
  base = float(input("Enter the base: "))
  exponent = float(input("Enter the exponent: "))
  print("Power =", math.pow(base, exponent))

elif choice == 5:
  angle = float(input("Enter angle in degrees: "))
  rad = math.radians(angle)
  print("Sin =", math.sin(rad))
  print("Cos =", math.cos(rad))
  print("Tan =", math.tan(rad))

elif choice == 6:
  num1 = int(input("Enter first number: "))
  num2 = int(input("Enter second number: "))
  print("GCD =", math.gcd(num1, num2))

elif choice == 7:
  num1 = int(input("Enter first number: "))
  num2 = int(input("Enter second number: "))
  print("LCM =", math.lcm(num1, num2))
```
